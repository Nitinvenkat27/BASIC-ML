```python
X = data.drop('medv', axis=1).values
y = data['medv'].values
```

```python
reg = linear_model.LinearRegression()
reg.fit(X, y)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_pr = reg.predict(X)

print(f'MAE = {mean_absolute_error(y, y_pr)}')
print(f'MSE = {mean_squared_error(y, y_pr)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y, y_pr))}')
print(f'R_2 = {r2_score(y, y_pr)}')
```
